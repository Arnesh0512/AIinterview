```python
from datetime import datetime
from collections import defaultdict
import math
```

```python

def normalize_and_rank(candidates_scores):

    # Step 1: Normalize all questions and accumulate final z-score
    results = defaultdict(lambda: {
        "final_normalized_score": 0.0,
        "latest_submission": None
    })

    for question_data in candidates_scores:
        if not question_data:
            continue

        raw_scores = [entry["raw_score"] for entry in question_data]
        n = len(raw_scores)

        mean = sum(raw_scores) / n
        variance = sum((x - mean) ** 2 for x in raw_scores) / n
        std_dev = math.sqrt(variance)

        # If all same score → z = 0
        if std_dev == 0:
            for entry in question_data:
                candidate_id = entry["candidate_id"]
                submitted_time = (
                    entry["submitted_at"]
                    if isinstance(entry["submitted_at"], datetime)
                    else datetime.fromisoformat(entry["submitted_at"])
                )
                prev_time = results[candidate_id]["latest_submission"]
                if prev_time is None or submitted_time > prev_time:
                    results[candidate_id]["latest_submission"] = submitted_time
            continue

        # Compute z-score and accumulate
        for entry in question_data:
            candidate_id = entry["candidate_id"]
            raw_score = entry["raw_score"]
            z_score = (raw_score - mean) / std_dev
            results[candidate_id]["final_normalized_score"] += z_score

            submitted_time = (
                entry["submitted_at"]
                if isinstance(entry["submitted_at"], datetime)
                else datetime.fromisoformat(entry["submitted_at"])
            )
            prev_time = results[candidate_id]["latest_submission"]
            if prev_time is None or submitted_time > prev_time:
                results[candidate_id]["latest_submission"] = submitted_time

    # Step 2: Create a list for ranking
    leaderboard = []
    for candidate_id, data in results.items():
        leaderboard.append({
            "candidate_id": candidate_id,
            "final_normalized_score": data["final_normalized_score"],
            "latest_submission": data["latest_submission"]
        })

    # Step 3: Sort leaderboard
    # 1) Higher final score first
    # 2) Tie → earlier submission wins
    leaderboard.sort(key=lambda x: (-x["final_normalized_score"], x["latest_submission"]))

    # Step 4: Assign rank and percentile
    total_candidates = len(leaderboard)
    current_rank = 1

    for i, candidate in enumerate(leaderboard):
        if i > 0 and (
                abs(candidate["final_normalized_score"] - leaderboard[i-1]["final_normalized_score"]) < 1e-9 and
                candidate["latest_submission"] == leaderboard[i - 1]["latest_submission"]
        ):
            candidate["rank"] = leaderboard[i - 1]["rank"]
        else:
            candidate["rank"] = current_rank
        current_rank += 1

        # Percentile: higher rank = higher percentile
        if total_candidates == 1:
            candidate["percentile"] = 100.0
        else:
            candidate["percentile"] = (
                (total_candidates - candidate["rank"]) /
                (total_candidates - 1)
            ) * 100

    return leaderboard

```

```python
import random
from datetime import datetime, timedelta
import json
def generate_candidates_scores(
    num_students=20,
    num_questions=10,
    min_score=0,
    max_score=10,
    start_time="2026-02-26T10:00:00"
):

    start = datetime.fromisoformat(start_time)
    current_time = start

    candidates_scores = []

    difficulties = ["hard", "medium", "easy"]

    for q in range(num_questions):

        difficulty = random.choice(difficulties)

        if difficulty == "hard":
            low = min_score
            high = max(min_score, int(max_score/3))

        elif difficulty == "medium":
            low = int(max_score/3)
            high = int(max_score*2/3)

        else:  # easy
            low = int(max_score*2/3)
            high = max_score

        question_data = []

        for s in range(1, num_students + 1):

            score = random.randint(low, high)

            question_data.append({
                "candidate_id": f"u{s}",
                "raw_score": score,
                "submitted_at": current_time.isoformat()
            })

            current_time += timedelta(seconds=30)

        candidates_scores.append(question_data)

    return candidates_scores



# Example usage
scores = generate_candidates_scores(
    num_students=20,
    num_questions=10,
    min_score=0,
    max_score=10
)


print("input_1 = [")

for qi, question in enumerate(scores):
    print(f"    [")

    for ci, entry in enumerate(question):
        if ci < len(question) - 1:
            print(f"        {entry},")
        else:
            print(f"        {entry}")

    if qi < len(scores) - 1:
        print("    ],")
    else:
        print("    ]")

print("]")


print("input_2 = [")

for qi, question in enumerate(scores):
    print(f"    [")

    for ci, entry in enumerate(question[0:15]):
        if ci < len(question) - 1:
            print(f"        {entry},")
        else:
            print(f"        {entry}")

    if qi < len(scores) - 1:
        print("    ],")
    else:
        print("    ]")

print("]")


for j in range(10):
    for i in [15,16,17,18,19]:
        scores[j][i]["raw_score"] = 0

        
print("input_3 = [")

for qi, question in enumerate(scores):
    print(f"    [")

    for ci, entry in enumerate(question):
        if ci < len(question) - 1:
            print(f"        {entry},")
        else:
            print(f"        {entry}")

    if qi < len(scores) - 1:
        print("    ],")
    else:
        print("    ]")

print("]")


for j in range(10):
    for i in [15,16,17,18,19]:
        scores[j][i]["raw_score"] = -1

print("input_4 = [")

for qi, question in enumerate(scores):
    print(f"    [")

    for ci, entry in enumerate(question):
        if ci < len(question) - 1:
            print(f"        {entry},")
        else:
            print(f"        {entry}")

    if qi < len(scores) - 1:
        print("    ],")
    else:
        print("    ]")

print("]")
```
